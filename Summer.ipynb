### 1. In how many cities Summer Olympics is held so far?

```python
sum_df = df.dropna()
sum_df
```

```python
len(sum_df.City.unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gmedals = sum_df[sum_df.Medal == 'Gold']
df_gmedals
```

```python
cgold = []
for Sport in df_gmedals.Sport.unique():
    cgold.append([Sport,len(df_gmedals[df_gmedals.Sport == Sport])])
cgold = pd.DataFrame(cgold,columns=['Sport','Gold Medals'])
cgold = cgold.sort_values(by = 'Gold Medals', ascending = False).head()
cgold.plot(x='Sport',y='Gold Medals',kind='bar',figsize=(7,5))
plt.title('GoldMedals')
print ("The Sport having most number of gold medals is : ", cgold.Sport[0])
```

```python
cgold
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
cmedals = []
for Sport in sum_df.Sport.unique():
    cmedals.append([Sport,len(sum_df[sum_df.Sport == Sport])])
cmedals = pd.DataFrame(cmedals, columns =['Sport','Medals'])
cmedals = cmedals.sort_values(by = 'Medals', ascending = False).head()
print("The Sport having most number of medals is : ",cmedals.Sport[0])
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df_women = df_gmedals[df_gmedals.Gender == 'Women']
df_women
```

```python
dfWomenGold = []

for sport in df_women.Sport.unique():
    
    dfWomenGold.append([sport, len(df_women[df_women.Sport == sport])])

dfWomenGold = pd.DataFrame(dfWomenGold, columns =['Sport','Gold Medals'])

dfWomenGold = dfWomenGold.sort_values(by = 'Gold Medals', ascending = False).head()

dfWomenGold
```

```python
dfWomenGold.plot(x = 'Sport', y = 'Gold Medals', kind = 'bar', figsize = (8,6))

print("The sport having most number of female gold medalists is : ", dfWomenGold.Sport[4])
```
